## Cleaning

Change of data type in the Max_BPM column

```python
df['Max_BPM'] = pd.to_numeric(df['Max_BPM'], errors='coerce')
df.info()
```

We can not fillin columns 'Gender' and 'Workout_Type' with average values. So I prefer to drop out lines with blank fields in this columns.
I think in real world age is super important for fitness data. So I decided not to fill it with average and delete lines with nans.

```python
df = df.dropna(subset = ['Age', 'Gender', 'Workout_Type'])
df.isna().sum()
```

If we delete all the lines with nans we will lose about 1/3 of dataset. Therefore I decided to fillin fileds with average values for each column.

```python
pd.set_option('future.no_silent_downcasting', True)
columns_to_fill = [
    "Avg_BPM", "Max_BPM", "Weight (kg)", "Height (m)", "Resting_BPM",
    "Session_Duration (hours)", "Calories_Burned", "Fat_Percentage",
    "Water_Intake (liters)", "Workout_Frequency (days/week)", "Experience_Level", "BMI"
]

for col in columns_to_fill:
    mean_value = df[col].mean()
    df.loc[:, col] = df[col].fillna(mean_value)
df.isna().sum()
```

```python
valid_workout_types = ["Strength", "Cardio", "HIIT", "Yoga"]
df.loc[~df['Workout_Type'].isin(valid_workout_types), 'Workout_Type'] = None
df = df.dropna(subset=['Workout_Type']).reset_index(drop=True)
```

## Hypothesis 1 checking

### Prove

BMI index is equal to weight divided by squared height.
Let's create a new column called Calculated_BMI.

### Data Transformation 1

Calculating BMI index according to formula: BMI = Weight / (Height) ^ 2

```python
df['Calculated_BMI'] = (df['Weight (kg)'] / (df['Height (m)'] ** 2))
subset = ['BMI', 'Calculated_BMI', 'Weight (kg)', 'Height (m)']
df[subset]
```

```python
plt.figure(figsize=(5,3))
subset = ['BMI', 'Calculated_BMI', 'Weight (kg)', 'Height (m)']
sns.heatmap(df[subset].corr(numeric_only = True), annot = True, cmap="Greys")
plt.title('Correlation');
```

Calculated_BMI is highly dependent on Weight and Height. 
BMI in dataset is not depended on Weight and Height.
There is no dependency between BMI and Calculated_BMI.

```python
fig = px.scatter(df, x = 'Weight (kg)', y = 'Calculated_BMI', color = 'Height (m)')
fig.show()
```
